```python
import numpy as np 
import scipy.stats as st 
import math
from statsmodels.stats.weightstats import ztest
```

Ví dụ 1: Theo tiêu chuẩn thì trọng lượng các gói mì chính được đóng trên một máy tự động là 453g. Kiểm tra ngẫu nhiên 81 gói ta thấy trọng lượng trung bình là 448g. Với mức ý nghĩa 0,05 có thể cho rằng trọng lượng các gói mì chính không đạt tiêu chuẩn hay không, biết rằng trọng lượng gói mì chính là biến ngẫu nhiên có phân phối chuẩn với độ lệch chuẩn là 36g? 

Xây dựng giả thuyết:

$$H_{0}: \mu_{} = 453$$
Giả thuyết thay thế:

$$H_{a}: \mu_{} \neq 453$$

```python
sample = np.random.normal(448, 36, 81)
alpha = .05
null = 453
#Kiểm định:
ttest_score, p_value = ztest(sample, value = null, alternative = 'two-sided')
if (p_value < alpha):
  print("Có bằng chứng để bác bỏ giả thuyết H0")
else: 
  print("chưa có bằng chứng để bác bỏ giả thuyết H0")
```

Ví dụ 2: Năng suất trung bình của một giống lúa ở các năm trước là 32,5 (tạ/ha). Năm nay người ta dựa vào phương pháp chăm sóc mới và hy vọng năng suất cao hơn năm trước. Điều tra trên 15 thửa ruộng thu được kết quả sau: 

  33,7    35,4    32,7   36,3    37,3   32,4   30,0  

  32,4   31,7   34,5   42,0    33,9   38,1   35,0    33,8   (tạ/ha)

  Với mức ý nghĩa 1% có thể chấp nhận niềm hy vọng đó hay không, biết rằng năng suất lúa là 1 biến ngẫu nhiên có phân phối chuẩn với phương sai là 10 (tạ/ha)

Xây dựng giả thuyết:

$$H_{0}: \mu_{} = 32.5$$
Giả thuyết thay thế:

$$H_{a}: \mu_{} > 32.5$$

```python
sample = [33.7, 35.4, 32.7, 36.3, 37.3, 32.4, 30.0, 
          32.4, 31.7, 34.5, 42.0, 33.9, 38.1, 35.0, 33.8]
alpha = .01
null = 32.5
#Kiểm định:
ttest_score, p_value = ztest(sample, value = null, alternative = 'larger')
print(p_value)
if (p_value < alpha):
  print("Có bằng chứng để bác bỏ giả thuyết H0")
else: 
  print("chưa có bằng chứng để bác bỏ giả thuyết H0")
```

Ví dụ 3: Theo tiêu chuẩn thì trọng lượng các bao gạo do một máy tự động đóng là 50g. Sau một thời gian hoạt động người ta nghi ngờ máy hoạt động không bình thường làm cho trọng lượng các bao gạo giảm đi. Lấy ngẫu nhiên 90 bao và cân thử thì thu được trọng lượng trung bình là 48.5 kg. Với mức ý nghĩa 5% có thể kết luận gì về điều nghi ngờ trên, biết rằng trọng lượng bao gạo là biễn ngẫu nhiên có phân phối chuẩn với độ lệch chuẩn là 2kg?

Xây dựng giả thuyết:

$$H_{0}: \mu_{} = 50$$
Giả thuyết thay thế:

$$H_{a}: \mu_{} < 50$$

```python
sample = np.random.normal(48.5, 2, 90)
alpha = .05
null = 50
#Kiểm định
ttest_score, p_value = ztest(sample, value = null, alternative = 'smaller')
print(p_value)
if (p_value < alpha):
  print("Có bằng chứng để bác bỏ giả thuyết H0")
else: 
  print("chưa có bằng chứng để bác bỏ giả thuyết H0")
```
