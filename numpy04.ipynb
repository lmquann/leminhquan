Câu 5: Thực hiện các yêu cầu sau:
- Xây dựng hàm tính giá trị trung bình
- Xây dựng hàm tính giá trị trung vị
- Xây dựng hàm tính giá trị yếu vị
- Xây dựng hàm tính phương sai
- Xây dựng hàm tính độ lệch chuẩn
- Hãy tạo ra ma trận A15x10 với các giá trị sinh bởi hàm arrange()
- Áp dụng 5 hàm ở trên cho từng hàng của ma trận A
- Áp dụng 5 hàm ở trên cho từng cột của ma trận A
Gợi ý: tham khảo hàm apply_along_axis() đã học ở bài trước.

```python
#Hàm tính giá trị trung bình:
def gtri_trbinh(A):
  S = 0
  for i in range(0,len(A)):
    S += A[i] 
  TB = S/len(A)
  return TB

#Hàm tính giá trị trung vị:
def gtri_trvi(A):
  if len(A) %2 != 0:
    Trvi = A[int((len(A)-1)/2)]
    return Trvi
  if len(A) %2 == 0:
    Trvi = (A[int(len(A)/2)] + A[int(len(A)/2 - 1)])/2
    return Trvi

#Hàm tính phương sai:
def tinhphsai(A):
  TB = gtri_trbinh(A)
  Tong = 0
  for i in range (0,len(A)):
    Tong += (A[i]- TB)**2
  phsai =  Tong/len(A)
  return phsai

#Hàm tính độ lệch chuẩn:
def dlchuan(A):
  phsai = tinhphsai(A)
  dlchuan = phsai**0.5
  return dlchuan
```

Hãy tạo ra ma trận A15x10 với các giá trị sinh bởi hàm arrange()

Áp dụng 5 hàm ở trên cho từng hàng của ma trận A

Áp dụng 5 hàm ở trên cho từng cột của ma trận A 
Gợi ý: tham khảo hàm apply_along_axis() đã học ở bài trước.

```python
A = np.arange(0,150)
A = A.reshape((10,15))
print("Ma trận A: \n", A)

#Áp dụng kết quả cho từng hàng:
print("Trung bình: \n",np.apply_along_axis(gtri_trbinh,1,A))
print("Trung vị: \n", np.apply_along_axis(gtri_trvi,1,A))
print("Phương sai: \n", np.apply_along_axis(tinhphsai,1,A))
print("Độ lệch chuẩn: \n", np.apply_along_axis(dlchuan,1,A))

#Áp dụng kết quả cho từng cột:
print("Trung bình: \n",np.apply_along_axis(gtri_trbinh,0,A))
print("Trung vị: \n", np.apply_along_axis(gtri_trvi,0,A))
print("Phương sai: \n", np.apply_along_axis(tinhphsai,0,A))
print("Độ lệch chuẩn: \n", np.apply_along_axis(dlchuan,0,A))
```
